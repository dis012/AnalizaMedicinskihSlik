## Medicinski kontekst: slikovna diagnostika možganskih bolezni

Tomografsko slikanje z magnetno resonanco (MR) se pogosto uporablja za spremljanje poteka možgansko-žilnih ali nevrodegenerativnih bolezni kot so multipla skleroza (MS), Alzheimerjeva bolezen in druge vrste demence. Potek teh bolezni je namreč povezan z morfološkimi spremembami v možganih, ki se pogosto pojavijo pred drugimi kliničnimi znaki. Morfološke spremembe se najpogosteje pojavijo v beli možganovini (WM; ang. *White Matter*), pri čemer gre lahko za reverzibilne ali nereverzibilne poškodbe oz. lezije WM. Kakršnekoli spremembe lezij ali pojav novih lezij so klinično pomemben parameter stanja omenjenih bolezni in se običajno ovrednotijo s pomočjo ročnega obrisovanja in merjenjem števila in volumna lezij. 

Stanje oz. napredovanje bolezni se odraža tudi v morfoloških spremembah zdravih možganskih struktur (npr. atrofija), zato se merjenje volumna običajno izvaja tudi na strukturah kot so WM, siva možganovina (GM; ang. *Gray Matter*) in likvoju (CSF; ang. *CerebroSpinal Fluid*). Pri posamezni MR preiskavi se zajame več različnih slik, kot so T1-, T2- in PD-utežene MR slike ([Slika 1a](#fig_mr_slikanje)), za zaznavo, obrisovanje in analizo zdravih možganskih struktur in lezij bele možganovine pa se uporabljajo vse omenjene MR slike. 

Ker je ročno obrisovanje 3D slik subjektivno, časovno zahtevno in nenatančno je bilo predlaganih kar nekaj avtomatskih postopkov razgradnje MR slik glave. Ti avtomatski postopki MR sliko razgradijo na osnovne možganske strukture (CSF, GM in WM) glede na določen kriterij oz. značilnico, ki je dovolj homogena (sivine, teksture, ...). Zanesljivost in natančnost teh postopkov zavisi predvsem od kakovosti zajetih MR slik, osnovni postopki razgradnje pa temeljijo na odvodih, upragovanju, razvrščanju ali pa na matematičnih, fizičnih ali statističnih modelih ([Slika 1b](#fig_mr_slikanje)). 

V zadnjem času so se za razgradnjo MR slik uveljavili postopki, ki temeljijo na rojenju sivinskih vrednosti z neparametričnim razvrščanjem. Eden od takih postopkov, ki ga boste za razgradnjo MR slik razvili pri nalogi tej nalogi je postopek $k$-povprečij (ang. $k$*-means*). Predno tovrstne postopke lahko uporabimo za potrebe slikovne diagnostike možganskih bolezni jih je potrebno kvantitativno ovrednotiti. To običajno naredimo s primerjavo avtomatsko pridobljenih obrisov z referenčnimi ročnimi obrisi ([Slika 1defg](#fig_mr_slikanje)), ki pa so pridobljeni s statistično analizo in združevanjem večih ročnih obrisov. Pri nalogi bodo dani referečni obrisi, s katerimi boste lahko kvantitativno vrednotili kakovost razgradnje s postopkom $k$-povprečij.

## Razgradnja 3D MR slik

Za razgradnjo zdravih in bolezenskih možganskih struktur iz MR slik glave se običajno uporablja več sekvenc (npr. T1-, T2- in PD-utežena ([Slika 1a](#fig_mr_slikanje)), ker se na posameznih sekvencah oz. MR slikah te strukture izražajo z različnimi, a značilnimi sivinskimi vrednostmi. Če so MR slike med seboj prostorsko poravnane, potem sivinske vrednosti več različnih MR slik skupaj tvorijo večdimenzionalen *prostor značilnic*. Večje možganske strukture imajo v MR slikah običajno približno konstantne sivinske vrednosti, zato večja zastopanost teh med seboj podobnih sivinskih vrednosti v prostoru značilnic oblikuje gruče ([Slika 1b](#fig_mr_slikanje)). Analiza gruč (ang. *cluster analysis*) v prostoru značilnic MR slik je eden od uveljavljenih načinov razgradnje MR slik glave. Eden od osnovnih postopkov za analizo gruč na osnovi katerega lahko izvedemo razgradnjo MR slik z neparametričnim razvrščanjem je postopek $k$-povprečij.

Naj bodo $I_{T1}(\boldsymbol{x})$, $I_{T2}(\boldsymbol{x})$ in $I_{PD}(\boldsymbol{x})$ 3D sivinske slike, $\boldsymbol{x} = [x_1, x_2, x_3]^{\mathrm{T}}$, ki predstavljajo med seboj prostorsko poravnane T1-, T2- in PD-utežene MR sekvence. Za neko poljubno točko $\boldsymbol{x}_j$ v MR sekvencah lahko zapišemo vektor sivinskih vrednosti $\boldsymbol{y}_j = \left[\, I_{T1}(\boldsymbol{x}_j),\, I_{T2}(\boldsymbol{x}_j),\, I_{PD}(\boldsymbol{x}_j)\,\right]^{\mathrm{T}}$, ki ima dimenzije $3 \times 1$. Množice vektorjev $\boldsymbol{y}_j,\, j=1,\ldots,N,$ predstavljajo značilnice na osnovi katerih bomo s postopkom $k$-povprečij in neparametričnim razvrščanjem načrtali razgradnjo možganskih struktur. 

### Postopek $k$-povprečij

Za dano množico značilnic $Y=(\boldsymbol{y_1}, \boldsymbol{y_2}, \ldots, \boldsymbol{y_N})$, kjer je vsaka značilnica $d$-dimenzionalni vektor, postopek $k$-povprečij razvrsti teh $N$ značilnic v $k$ različnih množic $\boldsymbol{S} = \{S_1, S_2, \ldots, S_k\}$ ($k \leq N$) tako, da pri tem minimizira vsoto kvadratov usrediščenih razdalj značilnic znotraj vsake množice:
\begin{equation*}
	E^2 = \underset{\boldsymbol{S}}{\arg\min} \sum_{i=1}^k \sum_{\boldsymbol{y_j}\in S_i} \|\boldsymbol{y_j} - \boldsymbol{\mu_i}\|^2\,,
\end{equation*}
kjer je $\boldsymbol{\mu_i}$ središče oz. povprečje značilnic v množici $S_i$, $\|\cdot\|$ pa $L_2$ vektorska norma. Postopek $k$-povprečij minimizira $E^2$ na iterativen način tako, da v vsakem koraku $t$ izvede naslednji dve operaciji:

<a id='eq_kmeans_step1'></a>
  **1. Razvrščanje značilnic:** Značilnico $\boldsymbol{y}_j$ razvrsti v tisto množico $S_{i^*}^{(t)}$, katere središče $\boldsymbol{\mu}_{i^*}^{(t)}$ je najbližje v smislu Evklidske razdalje:
\begin{equation}
	S_{i^*}^{(t)} = \{ \boldsymbol{y}_j : \|\boldsymbol{y}_j - \boldsymbol{\mu}_{i^*}^{(t)}\| \leq \|\boldsymbol{y}_j - \boldsymbol{\mu}_i^{(t)}\|, \;\forall i, 1 \leq i \leq k \}\,,
	\label{eq:kmeans_step1}
\end{equation}
kjer vsak $\boldsymbol{y}_j$ pripišemo natanko eni od množic $\boldsymbol{S}^{(t)}$.

<a id='eq_kmeans_step2'></a>
  **2. Posodabljanje središč:** Izračunaj nova središča $\boldsymbol{\mu}_i^{(t)}$ kot povprečja značilnic v množici $S_{i}^{(t)}$:
\begin{equation}
	\boldsymbol{\mu}_i^{(t+1)} = \frac{1}{| S_i^{(t)} |} \sum_{\boldsymbol{y}_j \in S_i^{(t)}} \boldsymbol{y}_j\,.
	\label{eq:kmeans_step2}
\end{equation}
Operaciji (1-2) ponavljamo dokler se položaji središč spreminjajo ($\boldsymbol{\mu}_i^{(t+1)} \neq \boldsymbol{\mu}_i^{(t)},\,\forall i$) ali do vnaprej določenega maksimalnega števila korakov $t_{max}$.

  **Določanje začetnih središč.** Pomemben vhodni parameter postopka $k$-povprečij so začetna središča $\boldsymbol{\mu}_i^{(0)},\,i=1,\ldots,k$. Pri nalogi boste uporabili razširjeni postopek $k$-povprečij++, ki določi tudi začetna središča z naslednjim postopkom:

  1. Naključno izberi eno izmed značilnic $\boldsymbol{y}_j$ in jo določi kot prvo središče; $\boldsymbol{\mu}_1^{(0)} = \boldsymbol{y}_j$.

  2. Izračunaj Evklidske razdalje vseh značilnic $\boldsymbol{y}_j$ do najbližjih že določenih središč $d_i(\boldsymbol{y}_j) = \| \boldsymbol{y}_j - \boldsymbol{\mu}_i^{(0)} \|, \forall i,\, 1 \leq i \leq i_m$.

  3. Določi novo središče $\boldsymbol{\mu}_{i_m+1}^{(0)}$ kot vsoto s kvadratom razdalje $d^2(\boldsymbol{y}_j)$ uteženih značilnic:
	\begin{equation*}
		\boldsymbol{\mu}_{i_m+1}^{(0)} = \frac{\sum_{\boldsymbol{y}_j \in S} d_i^2(\boldsymbol{y}_j) \, \cdot \boldsymbol{y}_j }{\sum_{\boldsymbol{y}_j \in S} d_i^2(\boldsymbol{y}_j)}\,.
	\end{equation*}

  4. Ponovi koraka 2 in 3, dokler ne določiš vseh $k$ začetnih središč.

### Razgradnja z neparametričnim razvrščanjem 
  
V nalogi boste uporabili postopek $k$-povprečij za analizo gruč v prostoru značilnic, določen s sivinskimi vrednostmi 3D slik $I_{T1}(\boldsymbol{x})$, $I_{T2}(\boldsymbol{x})$ in $I_{PD}(\boldsymbol{x})$. V postopku $k$-povprečij boste določili ustrezno število središč $k$ glede na pričakovano število velikih struktur v sliki; v slikah možganskih tkiv so to WM, GM in CSF. Značilnice $\boldsymbol{y}_j$ boste razvrstili v $k$ razredov glede na najmanjšo oddaljenost od $i$-tega središča $\boldsymbol{\mu}_i$, tj. minimalno $d_i(\boldsymbol{y}_j) = \| \boldsymbol{y}_j - \boldsymbol{\mu}_i \|$.Vsaka značilnica $\boldsymbol{y}_j$ dobi pripadajočo oznako razreda $z_j$:
\begin{equation}
	z_j = \underset{\forall i}{\arg\min}\, d_i(\boldsymbol{y}_j)\,.
	\label{eq:classification}
\end{equation}
Vsaki značilnici $\boldsymbol{y}_j$ v prostoru slike ustreza slikovni element s koordinatami $\boldsymbol{x}_j$, ki ga v sliki razgradnje $S$ zapišemo s pripadajočo oznako $S(\boldsymbol{x}_j) = z_j$. Postopek označevanja slikovnih elementov na osnovi analize gruč s postopkom $k$-povprečij je prikazan na [Sliki 2](#fig_razgradnja).

### Vrednotenje razgradnje MR slik

Zmogljivost postopkov razgradnje MR slik običajno ovrednotimo s stopnjo prekrivanja med referenčnimi obrisi možganskih struktur in avtomatsko razgradnjo teh struktur. Kot kvantitativna mera prekrivanja obrisov se najpogosteje uporablja Diceov koeficient podobnosti (DSC; ang. *Dice Similarity Coefficient*)
\begin{equation*}
	DSC_r = \frac{2\,|S_r \cap R_r|}{|S_r| + |R_r|}\,,\quad r = \{CSF, GM, WM\}
\end{equation*}
kjer $\cap$ predstavlja presek dveh regij, $|\cdot|$ pa velikost regije. $S_r$ je razgradnja pridobljena z avtomatskim postopkom, $R_r$ pa referenčna razgradnja.

kjer je vhodni parameter `iY` matrika značilnic dimenzij $N \times d$, `iK` pa število gruč $k$. Funkcija vrne matriko `oMu`, ki ima dimenzije $d \times k$ in ki predstavlja začetna središča. Določite tri začetna središča $\boldsymbol{\mu}_i^{(0)},\,i=1,2,3$ na matriki značilnic `Y`, ki ste jo ustvarili pri prejšnji nalogi kot $Y = \left[\, I_{T1}(\boldsymbol{x}),\, I_{T2}(\boldsymbol{x}),\, I_{PD}(\boldsymbol{x})\,\right]$ in ki ima dimenzije $N \times 3$.

````python
### BEGIN SOLUTION

def kMeansInit(y, k, type='k++'):
    
    y = np.asarray(y, dtype = 'float64')
    ylength = np.max(y.shape)

    if type == 'random':
        idx = np.random.choice(np.arange(ylength), size=k)
        mi = y[idx, :].transpose()

    elif type == 'k++':
        yj = y[np.random.randint(ylength), :] # nakljucno izberemo eno izmed znacilnic
        mi = [yj]

        for i in range(k-1):
            d = []
            for i in range(len(mi)):
                # izračunaj evklidske razdalje znacilnic do sredisc
                
                di = np.linalg.norm(y - mi[i], axis=1) 
                # dodaj na seznam
                d.append(di)

            dmin = np.min(d, axis=0) # evklidske razdalje znacilnic do najblizjih sredisc
            # dolocimo novo sredisce kot povprečje
            mi.append(
                np.sum(dmin**2*y.transpose(), axis=1)/np.sum(dmin**2)
            )
        mi = np.asarray(mi,dtype='float').transpose()
    return mi
### END SOLUTION
````

  kjer vhodni parametri `t1`, `t2` in `pd` predstavljajo 3D sivinske MR slike, zajete s T1-, T2- in PD-uteženo sekvenco. Vhodni parameter `mask` je binarna 3D slika maske možganskih struktur ([naloga 1](#naloga_1)). Vse vhodne slike imajo enake dimenzije. Uporabite funkciji `kMeansPP()` in `nonparClassification()` na značilnicah $Y$ iz področja maske za analizo gruč in neparametrično razvrščanje za določanje treh možganskih struktur CSF, GM in WM ($k=3$) na podlagi vseh treh vhodnih slik ($d=3$). Funkcija naj vrne 3D sliko oznak `oS`, ki ima enake dimenzije kot vhodne slike. 

   S pomočjo povprečnih T1-uteženih sivinskih vrednosti ([Slika 3](#fig_histogram)) v posamezni gruči določite standardne oznake  $L$ struktur CSF, GM in WM, ki so dane kot $z_{CSF}=1, z_{GM}=2$ in $z_{WM}=3$. Slikovne elemente v `oS`, ki ležijo zunaj maske `mask`, postavite na 0. Prikažite prečni prerez $x \times y \times 90$ izhodne slike oznak `oS` in uporabite barvno kodiranje sivinskih vrednosti (npr. z barvno lestvico `jet`).

````python
def mrBrainSegmentation(t1, t2, pd, mask, threshold = None):
    # BEGIN SOLUTION
    # to poenostavi med vajo
    #if pd is not None:
    #y.append(pd[mask])
    #if t2 is not None:
    #    y.append(t2[mask])
    #if t1 is not None:
    #    y.append(t1[mask])
    #else:
    #    print('razvrscanje v razrede CSF, GM, WM ne bo delovalo!')

    # število gruč (predznanje!)
    k = 3
    
    # zgradimo maske za vsako sliko
    It1 = t1[mask]
    It2 = t2[mask]
    Ipd = pd[mask]
    
    # uredimo v matriko znacilnic
    y = np.array([It1, It2, Ipd]).transpose()
    # nujno!
    y = np.asarray(y, dtype = 'float64')

    mi = kMeansPP(y, k) # analiza gruc      lahko y[::10, :]
    z, dmin = nonparClassification(y, mi) # neparametricno razvrscanje znacilnic

    ym = []
    # izracunamo povprecja sivinskih vrednosti v posamezni gruci
    for i in range(k):
        ym.append(np.mean(y[z == i, 0]))  # povprecja

    # glede na povprecja, dolocimo posamezne strukture
    icsf, igm, iwm = np.argsort(ym)

    # posameznim grucam dolocimo oznake
    L = np.zeros_like(z)
    L[z == icsf] = 1 # CSF
    L[z == igm] = 2 # GM
    L[z == iwm] = 3 # WM
    
    ## preskoci med vajo: poiscemo lezije
    #if threshold is not None:
    #    L[dmin > threshold] = LESIONS
    #
    
    # zgradimo izhodno sliko oznak
    S = np.zeros_like(mask).astype('float')
    S[mask>0] = L

    ### END SOLUTION
    return S
````
