1. Z 2D **togo preslikavo** $(t_x, t_y, \alpha) = (30 mm,\, -10 mm,\, 30^{\circ})$  preslikajte 2D mrežo kontrolnih točk, ki ste jih ustvarili pri Nalogi 2 v navodilih s funkcijo `getCubicBSpline2DGrid()`. 
  - Izrišite polje deformacij v obliki vektorskega polja z ukazom `quiver` v knjižici `matplotlib.pyplot`. Izrišite vektorje le za premik kontrolnih točk in ne vsake točke v koordinatnem sistemu.
  - Sliko šahovnice iz Naloge 2 v navodilih preslikajte z dobljenim poljem deformacij in jo prikažite.

```python
### BEGIN SOLUTION
#4 funkcija toge preslikave
#4 preslikava kontrolnih tock
#4 prikaz quiver
#4 preslikava tock sahovnice
#4 prikaz sahovnice

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interpn


def B0(u):
    # YOUR CODE HERE
    return (1 - u) ** 3.0 / 6.0


def B1(u):
    # YOUR CODE HERE
    return (3 * u ** 3 - 6 * u ** 2 + 4) / 6


def B2(u):
    # YOUR CODE HERE
    return (-3 * u ** 3 + 3 * u ** 2 + 3 * u + 1) / 6


def B3(u):
    # YOUR CODE HERE
    return u ** 3 / 6


B = [B0, B1, B2, B3]
    
def getCubicBSpline2DGrid(iImageSize, iStep):
    dy, dx = iImageSize
    if not isinstance(iStep, (list, tuple)):
        iStep = (iStep, iStep)
    oCPx, oCPy = np.meshgrid(
        np.arange(-iStep[0], (np.floor(dx/iStep[0])+3)*iStep[0], iStep[0]),
        np.arange(-iStep[1], (np.floor(dy/iStep[1])+3)*iStep[1], iStep[1]), 
        indexing = 'xy')
    return oCPx, oCPy


def getCubicBSpline2DDeformation(iImageSize, iCPx, iCPy, iStep):
    # YOUR CODE HERE
    dy, dx = iImageSize
    gx, gy = np.meshgrid(np.arange(dx), np.arange(dy), indexing='xy')
    gx = np.array(gx, dtype='float64')
    gy = np.array(gy, dtype='float64')

    oGx = np.zeros_like(gx)
    oGy = np.zeros_like(gy)
    i = np.floor(gx / iStep[0])
    j = np.floor(gy / iStep[1])
    u = gx / iStep[0] - i
    v = gy / iStep[1] - j

    for l in (0, 1, 2, 3):
        for m in (0, 1, 2, 3):
            oGx += B[l](u) * B[m](v) * iCPx[
                np.array(j, dtype='int64') + m,
                np.array(i, dtype='int64') + l]
            oGy += B[l](u) * B[m](v) * iCPy[
                np.array(j, dtype='int64') + m,
                np.array(i, dtype='int64') + l]
    return oGx, oGy


def deformImageBSpline2D(iImage, iCPx, iCPy, iStep):
    # YOUR CODE HERE
    dy, dx = iImage.shape
    oGx, oGy = getCubicBSpline2DDeformation(iImage.shape, iCPx, iCPy, iStep)

    gx, gy = np.meshgrid(np.arange(dx), np.arange(dy))
    oGy = 2 * gy - oGy
    oGx = 2 * gx - oGx
    # interpolacija
    oImage = interpn((np.arange(dx), np.arange(dy)),
                     iImage.astype('float'),
                     np.dstack((oGx, oGy)),
                     method='linear',
                     bounds_error=False, fill_value=0)

    return oImage


def getChessBoardImage(iImageSize, iArraySize=10, dtype='uint8'):
    dy = int(np.ceil(iImageSize[0] / iArraySize)) + 1
    dx = int(np.ceil(iImageSize[1] / iArraySize)) + 1

    A = [255 * np.ones(shape=(iArraySize, iArraySize)),
         np.zeros(shape=(iArraySize, iArraySize))]
    board = np.array(np.vstack([np.hstack([A[(i + j) % 2] \
                                           for i in range(dx)]) \
                                for j in range(dy)]), dtype=dtype)
    return board[:iImageSize[0], :iImageSize[1]]

# ___________________________________________________

def affine_transform(tx, ty, alpha, x, y):
    oShape = x.shape
    x = x.flatten()
    y = y.flatten()
    T_trans = np.array([[1, 0, tx],[0, 1, ty],[0, 0, 1]])
    T_rot = np.array([[np.cos(alpha), -np.sin(alpha), 0],[np.sin(alpha), np.cos(alpha), 0],[0, 0, 1]])

    mat = np.vstack((x, y, np.ones_like(x)))
    xy_a = np.dot(T_trans, np.dot(T_rot, mat))
    return xy_a[0].reshape(oShape), xy_a[1].reshape(oShape)

# ---------------------

# določi korak vzorčenja kontrolnih točk
iStep = (80, 80)
# ustvari kontrolne točke
oCPx, oCPy = getCubicBSpline2DGrid((400,400), iStep)
oCPx_a, oCPy_a = affine_transform(30, -10, 30 * np.pi / 180, oCPx, oCPy)  # 30 * np.pi / 180.

# Kam se preslikajo točke
# naredi izris
plt.close('all')
fig = plt.figure(figsize=(10, 10))    
ax = fig.add_subplot(111)
ax.quiver(oCPx, oCPy, oCPx_a-oCPx, oCPy_a-oCPy, - np.sqrt(oCPx_a**2 + oCPy_a**2),
          angles='xy', scale_units='xy',width=0.002, headwidth=8, headlength=10)  # berz scale=1 dobimo samo smer
ax.plot(oCPx, oCPy, marker='o', color='g', linewidth=1)
ax.plot(oCPx.transpose(), oCPy.transpose(), marker='o', color='g', linewidth=1)
ax.plot(oCPx_a, oCPy_a, marker='o', color='r', linewidth=1)
ax.plot(oCPx_a.transpose(), oCPy_a.transpose(), marker='o', color='r', linewidth=1)
#ax.set_xlim([-350,650])
#ax.set_ylim([-150,750])
ax.set_title('Opcija 1; z dejansko velikostjo premika; origin=lower')
plt.show()

# Kam se preslikajo točke
# naredi izris
plt.close('all')
fig = plt.figure(figsize=(10, 10))    
ax = fig.add_subplot(111)
ax.quiver(oCPx, oCPy, oCPx_a-oCPx, oCPy_a-oCPy, - np.sqrt(oCPx_a**2 + oCPy_a**2),
          angles='xy', scale_units='xy', scale=1, width=0.002, headwidth=8, headlength=10)  # berz scale=1 dobimo samo smer
ax.plot(oCPx, oCPy, marker='o', color='g', linewidth=1)
ax.plot(oCPx.transpose(), oCPy.transpose(), marker='o', color='g', linewidth=1)
ax.plot(oCPx_a, oCPy_a, marker='o', color='r', linewidth=1)
ax.plot(oCPx_a.transpose(), oCPy_a.transpose(), marker='o', color='r', linewidth=1)
ax.set_xlim([-350,650])
ax.set_ylim([750,-150])
ax.set_title('Opcija 1; z dejansko velikostjo premika')
plt.show()


# Kam se preslikajo tocke -- opcija 2; 
plt.close('all')
fig = plt.figure(figsize=(10, 10))    
ax = fig.add_subplot(111)
ax.quiver(oCPx, oCPy, np.subtract(oCPx_a,oCPx),-np.subtract(oCPy_a,oCPy), 
          width=0.002, headwidth=8, headlength=10)  # scale=1,
ax.plot(oCPx, oCPy, marker='o', color='g', linewidth=1)
ax.plot(oCPx.transpose(), oCPy.transpose(), marker='o', color='g', linewidth=1)
ax.plot(oCPx_a, oCPy_a, marker='o', color='r', linewidth=1)
ax.plot(oCPx_a.transpose(), oCPy_a.transpose(), marker='o', color='r', linewidth=1)
ax.set_xlim([-350,650])
ax.set_ylim([750,-150])
ax.set_title('Opcija 2; samo smer premika')
plt.show()

# preslikamo šahovnico
cbImage = getChessBoardImage((400, 400), 50)
de_x, de_y = getCubicBSpline2DDeformation(cbImage.shape, oCPx_a, oCPy_a, iStep)
x1CP = np.arange(0, 400, 1)
y1CP = np.arange(0, 400, 1)
X, Y = np.meshgrid(x1CP, y1CP)
# prikažemo preslikano šahovnico
cbImageDeformed = deformImageBSpline2D(cbImage, oCPx_a, oCPy_a, iStep)


f, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 3))
ax1.imshow(cbImage, cmap='gray')
ax2.imshow(cbImageDeformed, cmap='gray')
#ax2.set_xlim([-150,650])
#ax2.set_ylim([650,-150])
f.suptitle('Origin upper')
plt.show()


f, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 3))
ax1.imshow(cbImage, cmap='gray', origin='lower')
ax2.imshow(cbImageDeformed, cmap='gray', origin='lower')
#ax2.set_xlim([-150,650])
#ax2.set_ylim([650,-150])
f.suptitle('Origin lower')
plt.show()

### END SOLUTION
```
